# data preprocessing

```python
#ابتدا از ستون تایتل ، سال ها را حذف میکنیم
movies_df['title'] = movies_df.title.str.replace('(\(\d\d\d\d\))', '')
movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())
movies_df.head(10)
```

```python
#ستون ژانر را که به وسیله | جدا شده اند را به یک لیست تبدیل میکنیم
movies_df['genres'] = movies_df.genres.str.split('|')
movies_df.head()
```

```python
#در هر سطر ، لیست های مربوط به ژانر را چک میکنیم و اگر آن ژانر در لیست آن سطر انتخاب شده باشد به آن عدد 1 نسبت می دهیم و در غیر این صورت عدد صفر را نسبت میدهیم
for i, j in movies_df.iterrows():
    for k in j['genres']:
        movies_df.at[index, genre] = 1
movies_df = movies_df.fillna(0)
movies_df.head()
```

```python
movies_df['title']
```

```python
#حال به دلخواه یک سری ورودی با تایتل و امتیاز رندوم انتخاب میکنیم
userInput = [
            {'title':'Grumpier Old Men', 'rating':5},
            {'title':'Toy Story', 'rating':5},
            {'title':'Jumanji', 'rating':5},
            {'title':"No Game No Life: Zero", 'rating':1},
            {'title':'Bungo Stray Dogs: Dead Apple', 'rating':4}
         ] 
inputMovies = pd.DataFrame(userInput)
inputMovies
```

```python
#فیلم های ورودی را در دیتافریم اصلی پیدا میکنیم 
inputId = movies_df[movies_df['title'].isin(inputMovies['title'].tolist())]
inputId
```

```python
inputMovies = pd.merge(inputId, inputMovies)
inputMovies
```

```python
#برای محاسبات و تشکیل ماتریس ، به ستون تایتل و ژانر و ایدی فیلم نیازی نداریم
userGenreTable = inputMovies.drop('movieId', 1).drop('title', 1).drop('genres', 1)
userGenreTable
```

```python
#نمره هایی که در ابتدا به فیلم ها به دلخواه داده شد
inputMovies['rating']
```

```python
#حال باید ماتریس ساخته شده در بالا را در امتیاز داده شده به هر فیلم ضرب کنیم ، تا یک پروقایل از یوزرمان داشته باشیم
userProfile = userGenreTable.transpose().dot(inputMovies['rating'])
userProfile
#یوزر ما بیشترین امتیاز را به فیلم های کمدی و فانتزی داده است
```

```python
#حال ماتریس اصلی که شامل فیلم ها و ژانر هاشون هس رو تشکیل می دهیم
genreTable = movies_df.set_index(movies_df['movieId'])
genreTable = genreTable.drop('movieId', 1).drop('title', 1).drop('genres', 1)
genreTable.head()
```

```python
#سپس برای تشکیل جدول پیشنهادی ، ماتریس پروفایل یوزر را در ماتریس بالا(فیلم و ژانرهاشون) ضرب میکنیم و ستون های آن را جمع می کنیم 
recommendationTable_df = ((genreTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df.head()
```

```python
recommendationTable_df = recommendationTable_df.sort_values(ascending=False)
recommendationTable_df.head()
```
